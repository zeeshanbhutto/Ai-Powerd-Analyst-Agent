```python
import faiss
import numpy as np
import ollama

class EmbeddingStore:
    def __init__(self):
        """Initialize FAISS index dynamically based on the first embedding size"""
        sample_embedding = self.generate_embedding("test text")
        self.dimension = len(sample_embedding)  # Set correct dimension
        self.index = faiss.IndexFlatL2(self.dimension)  # Use L2 distance
        self.text_chunks = []  # To store text alongside embeddings

    def generate_embedding(self, text):
        """Generate embedding using Ollama"""
        try:
            response = ollama.embeddings(model="gemma2:2b", prompt=text)
            return np.array(response['embedding'], dtype=np.float32)  # Extract embedding list
        except Exception as e:
            print(f"‚ùå Error generating embedding: {e}")
            return np.zeros(self.dimension, dtype=np.float32)  # Return zero vector on failure

    def store_embedding(self, text):
        """Generate and store embedding in FAISS index"""
        embedding = self.generate_embedding(text).reshape(1, -1)  # Reshape for FAISS
        if embedding.shape[1] == self.dimension:  # Ensure correct shape
            self.index.add(embedding)  
            self.text_chunks.append(text)
            print("‚úÖ Embedding stored successfully!")
        else:
            print("‚ùå Mismatched embedding dimensions!")

    def retrieve_similar(self, query_text, top_k=5):
        """Retrieve similar embeddings from FAISS"""
        query_embedding = self.generate_embedding(query_text).reshape(1, -1)
        if query_embedding.shape[1] == self.dimension:
            distances, indices = self.index.search(query_embedding, top_k)
            results = [self.text_chunks[i] for i in indices[0] if i < len(self.text_chunks)]
            return results
        else:
            print("‚ùå Mismatched embedding dimensions!")
            return []

# ‚úÖ Running the embedding store
if __name__ == "__main__":
    store = EmbeddingStore()

    print("‚ö° Storing embedding...")
    store.store_embedding("This is a test sentence for embeddings.")

    print("üîç Retrieving similar results...")
    results = store.retrieve_similar("Find similar text.")
    print("üìå Similar results:", results)
```
